### Activation Function Implement in Python

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sigmoid Activation Function
def Sigmoid(x):
    return(1/(1+np.exp(-x)))
```

```python
# tanH Activation Function
# We can also use np.tanh(x)
def tanH(x):
    return((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
```

```python
# ReLu Activation Function
def ReLu(x):
    return(np.maximum(0,x))
```

```python
# Leaky ReLu Activation Function
# We can also use Leaky_ReLu = lambda a: lambda x: np.where(x>0, x, alpha*x)
# To call this function, use: Leaky_ReLu(0.5)(x)
def Leaky_ReLu(x, alpha=0.5):
    return(np.where(x>0, x, alpha*x))
```

```python
# Data Visualization
x=np.linspace(-2, 2, 200)
fig, ax = plt.subplots(figsize=(20,10))

plt.plot(x, Sigmoid(x))
plt.plot(x, tanH(x))
plt.plot(x, ReLu(x))
plt.plot(x, Leaky_ReLu(x))

plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphs for Different Activation Functions')
plt.grid(which='major', linestyle='--', linewidth=2, alpha=0.5)
plt.legend(['Sigmoid','tanH','ReLu','Leaky ReLu'], loc=1, prop={'size': 15})
plt.show()
```

### Gradient Descent Implement in Python

```python
# Function x
1/2*x**2
```

```python
# Helper function
def grad_f(x):
    return(x)

# Main function
def grad_descent(x_0, iterations, stepsize):
    
    # Create an array to store the results
    x=x_0
    res=[x]
    
    # For loop to find the minimum value
    for i in range(iterations):
        grad=grad_f(x)

        if grad<0:
            dx=stepsize*grad
        elif grad>0:
            dx=-stepsize*grad
        else:
            dx=0

        # New x
        x=x+dx
        res.append(x)
    
    return(res)

# Call function
x_0=2
iterations=100
stepsize=0.01
xs=grad_descent(x_0, iterations, stepsize)

# Plot the result
fig, ax = plt.subplots(figsize=(20,10))
plt.plot(xs,'.-')
plt.ylabel('x')
plt.xlabel('iteration')
plt.show()
```
